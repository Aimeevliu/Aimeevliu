```python
import pandas as pd
import matplotlib.pyplot as plt

# 設置字體和樣式
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.style.use('ggplot')

# 讀取 Excel 檔案，並跳過前 7 行
data = pd.read_excel('11308/index-2.xlsx', skiprows=7)

# 選擇需要的行範圍，並去掉第一欄
data = data.iloc[6:28, 1:]

# 刪除包含 '113年' 的整行
data = data[~data.eq('113年').any(axis=1)]

# 自動生成欄位名稱
data.columns = [f"Col{i+1}" for i in range(len(data.columns))]
print(data.head())

# 設置繪圖大小
plt.figure(figsize=(10, 5))

# 繪製折線圖
plt.plot(data['Col35'], data['Col2'], marker='*', linestyle='-')

# 設置標題和座標軸標籤
plt.title('金融機構存款餘額')
plt.xlabel('日期')
plt.ylabel('數值')

# 顯示圖表
plt.show()
```
