```python
# rebase 股價繪圖 #'^DJI', '^TWII','^HSI','^N225','^STI',^GSPC(標普500),
import ffn 
prices = ffn.get(['NVDA','AAPL'], start='2024-01-01')
%matplotlib inline
prices.rebase().plot() # rebase 股價繪圖
prices.to_drawdown_series().plot() # Max Drawdown 最大虧損
```

```python
# 統計報表
stats = prices.calc_stats()
stats.display()
```

```python
# 下載股票歷史價格
prices = ffn.get('^DJI','INTC','NVDA', start='2020-01-01')

# 計算日回報
returns = prices.to_returns().dropna()

# 計算績效摘要
perf = returns.calc_stats()
perf.display()
```
