```python
import requests
import pandas as pd
import json
import datetime as dt
from IPython.display import display, HTML

data = []
url = "https://api.cnyes.com/media/api/v1/newslist/category/headline"
payload = {
    'page': 1,
    'limit': 30,
    'isCategoryHeadline': 1,
    'startAt': int((dt.datetime.today() - dt.timedelta(days=10)).timestamp()),
    'endAt': int(dt.datetime.today().timestamp())
}

# 請求第一頁資料
res = requests.get(url, params=payload)
jd = json.loads(res.text)
df = pd.DataFrame(jd['items']['data'])

# 保存原始的 newsId 值，用來生成超連結
df['original_newsId'] = df['newsId']

# 假設使用 original_newsId 的變化來生成不同的時間
df['newsId'] = df['original_newsId'].apply(lambda x: (dt.datetime.today() - dt.timedelta(seconds=int(x) % 10000)).strftime('%Y-%m-%d %H:%M:%S'))

# 生成新聞鏈接，並使用原始的 newsId (timestamp) 生成對應的超連結
df['link'] = df['original_newsId'].apply(lambda x: f'<a href="https://news.cnyes.com/news/id/{x}">點擊查看新聞</a>')

# 添加第一頁新聞
data.append(df)

# 從第二頁開始抓取
for i in range(2, jd['items']['last_page'] + 1):
    payload['page'] = i
    res = requests.get(url, params=payload)
    jd = json.loads(res.text)
    df = pd.DataFrame(jd['items']['data'])

    # 保存原始的 newsId 值，用來生成超連結
    df['original_newsId'] = df['newsId']

    # 假設使用 original_newsId 的變化來生成不同的時間
    df['newsId'] = df['original_newsId'].apply(lambda x: (dt.datetime.today() - dt.timedelta(seconds=int(x) % 10000)).strftime('%Y-%m-%d %H:%M:%S'))

    # 生成新聞鏈接，並使用原始的 newsId (timestamp) 生成對應的超連結
    df['link'] = df['original_newsId'].apply(lambda x: f'<a href="https://news.cnyes.com/news/id/{x}">點擊查看新聞</a>')

    # 將結果加入到 data 列表中
    data.append(df)

# 合併所有頁數的數據
df = pd.concat(data, ignore_index=True)

# 選取需要的欄位，顯示 newsId（日期格式）和對應的超連結
df = df[['newsId', 'title', 'summary', 'link']]

# 將 DataFrame 轉換為 HTML 格式，並使用 IPython 的 display 函數顯示
display(HTML(df.to_html(escape=False)))  # escape=False 保留超連結的功能
```
