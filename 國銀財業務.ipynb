```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # 用於位移計算

# 設置字體和樣式
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.style.use('ggplot')

# 讀取 Excel 檔案，並跳過前 7 行
data = pd.read_excel('11309.xlsx', header=1)

# 選擇需要的行範圍，並去掉第一欄
data = data.iloc[1:11, 0:8]

# 自動生成欄位名稱
data.columns = [f"Col{i+1}" for i in range(len(data.columns))]

# 將 X 軸欄位（Col1）轉換為字串，並填充 NaN 值
data['Col1'] = data['Col1'].fillna('').astype(str)

# 檢視資料
print(data.head())

# 設置繪圖大小
plt.figure(figsize=(8, 6))

# 定義 X 軸資料和 Y 軸資料
x = np.arange(len(data['Col1']))  # 使用數值索引以便位移
y1 = data['Col3']
y2 = data['Col5']
y3 = data['Col7']
width = 0.3  # 設定柱狀圖的寬度

# 繪製柱狀圖
plt.bar(x - width/2, y1, width=width, label='Col3')  # 垂直長條圖 1
plt.bar(x + width/2, y2, width=width, label='Col5')  # 垂直長條圖 2
# 設置 X 軸標籤和旋轉角度
plt.xticks(x, data['Col1'], rotation=45)  # 將文字旋轉 45 度

# 設置標題與軸標籤
plt.title('資料柱狀圖')
plt.xlabel('Col1')
plt.ylabel('數值')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.show()
```
