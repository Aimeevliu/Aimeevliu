```python
import httpx
from bs4 import BeautifulSoup as bs
import pandas as pd
from io import StringIO 

url = "https://www.sitca.org.tw/ROC/Industry/IN2411.aspx"
data = []

# 選擇所有 type 為 hidden 的 input 標籤，將每組 name 和 value 加入 data 列表
for ele in soup.select('input[type="hidden"]'):
    name = ele.get('name')
    value = ele.get('value', '')  # 默認值設置為空字符串，防止沒有 value 的情況
    if name:  # 確保 name 存在
        data.append((name, value))

# 將 hidden input 的 name 和 value 加入 payload 字典
payload = {key: value for key, value in data}
payload.update({
    'ctl00$ContentPlaceHolder1$ddlQ_YEAR': 202408,
    'ctl00$ContentPlaceHolder1$BtnQuery.x': 24,
    'ctl00$ContentPlaceHolder1$BtnQuery.y': 10
})

# 發送 POST 請求
res = httpx.post(url, data=payload, verify=False)
soup = bs(res.text, 'lxml') 
tb = soup.select('table')[0] #第一個是0
df = pd.read_html(StringIO(tb.prettify())) # 只有 read_html 要加[0]
df[0].iloc[4:]
```
