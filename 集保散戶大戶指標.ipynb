```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from io import StringIO

url = "https://www.tdcc.com.tw/portal/zh/smWeb/qryStock"
headers = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36"
}
payload = {
    "scaDate": "",
    "sqlMethod": "StockNo",
    "stockNo": "2330",
    "stockName": ""
}
rs = requests.session()#模擬與瀏覽器可儲存cookie的功能
res = rs.get(url, headers = headers)
soup = bs(res.text, "lxml") #解析網頁原始程式碼
inp = soup.select('input[type="hidden"]')  #取出網頁中的隱藏參數

for ele in inp:
    payload[ele['name']] = ele['value'] #將隱藏參數轉成傳送用的參數
    if ele['name'].find('firDate') > -1:
        payload['scaDate'] = ele['value'] #將firDate、scaDate欄位的數值設定成一樣的內容
res = rs.post(url, data = payload, headers = headers) #改用post再傳送一次參數
soup = bs(res.text, "lxml")
tb = soup.select('.table') #class使用 "." 開頭搭配 "table" 名稱
df = pd.read_html(StringIO(tb[0].prettify()))[0] #將資料轉成表格帶入pandas的DataFrame
df1 = pd.DataFrame([{
    "400張以下人數":df.iloc[:11, 2].sum(),
    "400張以下比例":df.iloc[:11, 4].sum(),
    "400張以上人數":df.iloc[11:15, 2].sum(),
    "400張以上比例":df.iloc[11:15, 4].sum()
}])
df1
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from io import StringIO
import datetime as dt

url = "https://www.tdcc.com.tw/portal/zh/smWeb/qryStock"
headers = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36"
}
res = requests.get(url, headers = headers)
soup = bs(res.text, "lxml") #解析網頁原始程式碼
dates = list(map(lambda x : x.text, soup.select('option') ))

payload = {
    "scaDate": "",
    "sqlMethod": "StockNo",
    "stockNo": "006208",
    "stockName": ""
}
data = []
for date in dates:
    rs = requests.session()#模擬與瀏覽器可儲存cookie的功能
    res = rs.get(url, headers = headers)
    soup = bs(res.text, "lxml") #解析網頁原始程式碼
    inp = soup.select('input[type="hidden"]')  #取出網頁中的隱藏參數
    
    for ele in inp:
        payload[ele['name']] = ele['value'] #將隱藏參數轉成傳送用的參數
    payload['firDate'] = date
    payload['scaDate'] = date
    
    res = rs.post(url, data = payload, headers = headers) #改用post再傳送一次參數
    soup = bs(res.text, "lxml")
    tb = soup.select('.table') #class使用 "." 開頭搭配 "table" 名稱
    df = pd.read_html(StringIO(tb[0].prettify()))[0] #將資料轉成表格帶入pandas的DataFrame

    #建立表格資料
    df1 = pd.DataFrame([{
        "日期":dt.datetime.strptime(date, '%Y%m%d'),
        "400張以下人數":df.iloc[:11, 2].sum(),
        "400張以下比例":df.iloc[:11, 4].sum(),
        "400張以上人數":df.iloc[11:15, 2].sum(),
        "400張以上比例":df.iloc[11:15, 4].sum()
    }])
    print(date)
    data.append(df1) #取得所有表格
df = pd.concat(data, ignore_index = True) #合併所有表格
df
```

```python
import yfinance as yfin
pd.set_option("display.max_rows", 250) #設定顯示全部row的資料內容前250筆
#pd.set_option("display.max_columns", 250) #設定顯示全部column的資料內容前250筆
date = df['日期'].iloc[-1].strftime("%Y-%m-%d")
df1 = yfin.download("006208.tw", date)
#pd.merge(table1, table2, 
#         left_on = "table1欄位依據",
#         right_on = "table2欄位依據",
#         how = "合併方法" inter:左右兩邊都有的再合併, outer:左右兩邊都不一定有的合併
#)
df2 = pd.merge(df1, df, left_on = df1.index, right_on = df['日期'], how = 'outer') #表格合併
df2
```

```python
#df2['400張以下人數'].fillna(method = 'ffill') #舊寫法未來要移除
df2[['400張以下人數', '400張以下比例', '400張以上人數', '400張以上比例']] = df2[['400張以下人數', '400張以下比例', '400張以上人數', '400張以上比例']].ffill()
df2
```

```python
import numpy as np
df2['400張以下人數_s'] = np.full(len(df2['400張以下人數']), np.nan)#建立一個新的欄位並設定初始值為nan
df2['400張以上人數_s'] = np.full(len(df2['400張以上人數']), np.nan)#建立一個新的欄位並設定初始值為nan

#當本周的人數小於下一周的人數，表示下一周人數增加，設定為1
df2['400張以下人數_s'] = np.where(df2['400張以下人數'] > df2['400張以下人數'].shift(), 1, df2['400張以下人數_s'])

#當本周的人數大於下一周的人數，表示下一周人數減少，設定為-1
df2['400張以下人數_s'] = np.where(df2['400張以下人數'] < df2['400張以下人數'].shift(), -1, df2['400張以下人數_s'])

#當本周的人數小於下一周的人數，表示下一周人數增加，設定為1
df2['400張以上人數_s'] = np.where(df2['400張以上人數'] > df2['400張以上人數'].shift(), 1, df2['400張以上人數_s'])

#當本周的人數大於下一周的人數，表示下一周人數減少，設定為-1
df2['400張以上人數_s'] = np.where(df2['400張以上人數'] < df2['400張以上人數'].shift(), -1, df2['400張以上人數_s'])

df2[['400張以下人數_s', '400張以上人數_s']] = df2[['400張以下人數_s', '400張以上人數_s']].ffill()
df2
```

```python
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick2_ochl
plt.rcParams['font.family'] = 'Arial Unicode MS' #mac os: Arial Unicode MS

df3 = df2.tail(75).reset_index(drop = True).copy() #只用最新的75筆資料
fig, ax = plt.subplots(3, 1, figsize = (10, 4))
candlestick2_ochl(ax[0],
                 df3['Open'],
                 df3['Close'],
                 df3['High'],
                 df3['Low'],
                 width = 0.8,
                 colorup = 'r',
                 colordown = 'g',
                 alpha = 0.5)

for idx, row in df3.iterrows(): #繪製散戶長條圖
    if row['400張以下人數_s'] == 1:
        ax[1].bar(idx, height = row['400張以下人數'], color = 'r')
    elif row['400張以下人數_s'] == -1:
        ax[1].bar(idx, height = row['400張以下人數'], color = 'g')
ax[1].yaxis.major.formatter.set_scientific(False) #關閉科學記號

for idx, row in df3.iterrows(): #繪製大戶長條圖
    if row['400張以上人數_s'] == 1:
        ax[2].bar(idx, height = row['400張以上人數'], color = 'r')
    elif row['400張以上人數_s'] == -1:
        ax[2].bar(idx, height = row['400張以上人數'], color = 'g')


ax[0].set_title('006208集保股權')
ax[0].set_ylabel('股價')
ax[0].set_xticks([]) #將X軸刻度關閉
ax[1].set_xticks([]) #將X軸刻度關閉
datelist = df3['key_0'].dt.strftime('%Y/%m/%d')#將日期轉字串
ax[2].set_xticks(range(0, len(datelist), 10)) #設定x軸刻度為序列資料
ax[2].set_xticklabels(datelist[::10], rotation= 30) #將X軸刻度為序列資料替換成日期
ax[1].legend(['散戶人數'])
ax[1].set_ylabel('散戶人數')
ax[2].legend(['大戶人數'])
ax[2].set_ylabel('大戶人數')
ax[2].set_xlabel('日期')
plt.subplots_adjust(hspace = 0)
fig.savefig("006208集保股權.png",
            dpi = 200,
            bbox_inches = 'tight')
```
