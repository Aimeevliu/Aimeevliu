```python
import requests
import pandas as pd
from io import StringIO
import yfinance as yf

data = []  # Initialize as a list to store DataFrames

for i in [1, 4]:
    url = "https://isin.twse.com.tw/isin/class_main.jsp?issuetype={}".format(i)  # 使用 format 函數將 i 代入 URL 中的 {} 區域，動態生成不同的網址
    res = requests.get(url)
    df = pd.read_html(StringIO(res.text), header = 0)[0]  # Set header row
    data.append(df)  # Append DataFrame to the list
df = pd.concat(data, ignore_index=True)

data = []
for idx , row in df.iterrows():
        if row['市場別'] == '上市':
            data.append(str(row['有價證券代號']) + '.tw')
        elif row['市場別'] == '上櫃':
            data.append(str(row['有價證券代號']) + '.two')
df = yf.download(data,"2024-01-01")
df

##	1.	for idx, row in df.iterrows()：這是 Pandas 的一個迴圈方法，會遍歷 df 這個 DataFrame 的每一行資料。idx 是當前行的索引（行號），而 row 是當前行的資料（以 Series 形式表示）。
##	2.	if row['市場別'] == '上市'：這裡檢查當前行的「市場別」欄位是否為「上市」。
##	•	如果「市場別」是「上市」，那麼會將當前行的「有價證券代號」欄位值轉為字串，並加上「.tw」這個字尾後綴，表示它是上市股票的代號格式，然後將結果添加到 data 列表中。
##	3.	elif row['市場別'] == '上櫃'：如果「市場別」是「上櫃」，那麼會將「有價證券代號」加上「.two」這個後綴，表示它是上櫃股票的代號格式，並且同樣將結果添加到 data 列表中。
```

```python
# 抓取個股歷史資料

import requests
import pandas as pd
import json
import datetime as dt
from dateutil.relativedelta import relativedelta
import time
import sqlite3 as lite #資料庫套件

#倒數計時

def countdown(t):
    while t > -1:
        min, sec = divmod(t, 60)
        if t > 0:
            print("{:02d}:{:02d}".format(min, sec), end='\r')  # 正確格式化 min 和 sec
        elif t == 0:
            print("{:02d}:{:02d}".format(min, sec), end='\n')  # 倒數結束時換行
        
        t -= 1
        time.sleep(1)

url = "https://www.twse.com.tw/rwd/zh/afterTrading/MI_INDEX"
payload = {
         'date' : '20241018',
        'type' : 'ALLBUT0999',
        'response' : 'json',

}
data = []
#加入日期
base = dt.datetime.today() - relativedelta(years = 0, months = 0, days = 15)  #設定從今天往前抓取的日期範圍
while base < dt.datetime.today():
    payload['date'] = base.strftime("%Y%m%d")
    print(payload['date']) #列印資料進度
    res = requests.get(url, params = payload)
    jd = json.loads(res.text)
    if jd['stat' ] == 'OK': #當交易日stat資料 ＝OK
        td = jd['tables'][8]
        df = pd.DataFrame(td['data'], columns = td['fields'])
        df.insert(0, "日期", jd['date']) #數字 0 表示將新欄位插入為第一個欄位。
        df = df.iloc[:5, :4] #取前五行，前四列
        df = df.apply(lambda x: x.str.replace(',','')) #把逗號取代掉
        data.append(df)
        countdown(15) #每一次抓取後停留15秒
        
    base += relativedelta(days = 1) #以天為單位進行累加
df = pd.concat(data, ignore_index = True)

with lite.connect("每日行情.db") as con:
    df.to_sql("個股每日行情表", con = con, if_exists = 'replace') #如果有存在就覆蓋
df
```
