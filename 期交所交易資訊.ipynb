```python
import requests
import pandas as pd
from io import StringIO
url = "https://www.twse.com.tw/rwd/zh/fund/BFI82U"
payload = {
    "response":"html"
}
payload["response"] = "csv"
res = requests.get(url, params = payload)
df = pd.read_csv(StringIO(res.text), header = 1)
df.iloc[:6, :4].apply(lambda x:x.str.replace(",", ""))
```

```python
import requests
import pandas as pd
import json
url = "https://www.twse.com.tw/rwd/zh/fund/BFI82U"
payload = {
    "response":"html"
}
payload["response"] = "json"
res = requests.get(url, params = payload)
jd = json.loads(res.text) #字串轉換成字典
df = pd.DataFrame(jd['data'], columns = jd['fields']) #取值
df = df.apply(lambda x:x.str.replace(",", "")) #把 "," 拿掉後

df.to_csv("123.csv", encoding = "big5", index = False) # 轉成csv檔，名稱為123.csv，中文編碼 （big5、big5hkscs、utf_8_sig)，拿掉索引值
```
