```python
import yfinance as yfin
import pandas as pd
import numpy as np
from scipy.signal import argrelextrema

# 下載多支股票的資料
tickers = ["00679B.TWO", "00687B.TWO", "00795B.TWO", "00779B.TWO", "00853B.TWO", "00696B.TWO", "00764B.TWO", "00857B.TWO"]

stock_df = yfin.download(tickers, start="2024-05-01", end="2024-10-23", interval="1d", group_by="ticker")

# 設定窗口大小來判斷相對低點
window = 5  # 調整窗口大小來決定波段的敏感度

for ticker in tickers:
    if ticker in stock_df.columns.get_level_values(0):
        close_prices = stock_df[(ticker, 'Close')]

        # 找出波段低點（局部極小值）
        local_minima_indices = argrelextrema(close_prices.values, np.less, order=window)[0]

        # 抓取相對低點的日期和對應收盤價
        local_minima_dates = close_prices.index[local_minima_indices]
        local_minima_values = close_prices.iloc[local_minima_indices]

        # 提取最新的收盤價
        latest_price = close_prices.iloc[-1]
        latest_date = close_prices.index[-1]

        print(f"\n{ticker} 的波段低點與最新收盤價比較:")
        print(f"最新收盤價: {latest_price} (日期: {latest_date})")

        for date, value in zip(local_minima_dates, local_minima_values):
            difference = latest_price - value
            days_since = (latest_date - date).days  # 計算距離最新日期的天數
            print(f"波段低點日期: {date}, 收盤價: {value}, 與最新收盤價差異: {difference:.2f}, 距離低點天數: {days_since} 天")
```
