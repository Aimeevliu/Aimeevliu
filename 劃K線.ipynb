### data = yf.download(tickers, start=None, end=None, interval="1d", group_by='ticker', auto_adjust=False, prepost=False, threads=True, proxy=None)

- tickers：一個字符串或字符串列表，代表想要下載數據的股票代碼。例如 "AAPL" 或 ["AAPL", "MSFT", "GOOG"]。
- start 和 end：分別代表數據下載的起始和結束日期。它們應該是字符串（例如 "YYYY-MM-DD"）或 datetime 對象。
- interval：數據的時間間隔。可選值包括 "1m"（一分鐘）、"1h"（一小時）、"1d"（一天）、"1wk"（一周）和 "1mo"（一個月）。
- group_by：決定如何組織返回的數據。通常設置為 'ticker'。
- auto_adjust：是否自動調整數據，以反映股息和股票分割。
- prepost：是否包括盤前和盤後的交易數據。
- threads：是否使用多線程來加速下載。
- proxy：用於下載數據的代理服務器。

```python
# 畫出當天股價走勢
import yfinance as yfin
from mplfinance.original_flavor import candlestick2_ohlc, volume_overlay
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'Arial Unicode MS'
#[台股："^TWII",道瓊：^DJI]

df = yfin.download(["V"], start="2024-10-28", end="2024-10-30", interval="5m", group_by="ticker")
# plt.subplots(nrow, ncol, figsize) ,nrow:水平數量， ncol:垂直數量
fig, ax = plt.subplots(2, 1, figsize = (14, 4)) #框架大小


candlestick2_ohlc(ax[0],
                  df["Open"],
                  df["High"],
                  df["Low"],
                  df["Close"],
                  colorup = "r",
                  colordown = "g",
                  width = 0.5,
                  alpha = 0.5),
volume_overlay(ax[1],
                  df["Open"],
                  df["Close"],
                  df["Volume"],
                  colorup = "r",
                  colordown = "g",
                  width = 0.5,
                  alpha = 0.5, 
 )

dates = df.index.strftime("%Y/%m/%d %H:%M")#日期時間轉字
#ax[0].set_xticks(range(0, len(dates), 10)) #設定X軸刻度編號
#ax[0].set_xticklabels(dates[::10],rotation = 30) #繪製X軸日期，旋轉30度
#ax[0].legend() #開啟圖例

ax[0].xaxis.set_visible(False)
ax[1].set_xticks(range(0, len(dates), max(1, len(dates) // 10)))
ax[1].set_xticklabels(dates[::max(1, len(dates) // 10)], rotation=30)
# ax[1].yaxis.major.formatter.set_scientific(False) #關閉科學記號
# plt.rcParams['axes.unicode_minus'] = False #解決座標軸負數的負號顯示問題

plt.style.use('bmh')
plt.subplots_adjust(hspace = 0) #設定兩張圖的距離
plt.savefig("V股價走勢.png",
          format = "png",
            dpi = 200,
            bbox_inches = 'tight', transparent = True) #存成圖片
```
